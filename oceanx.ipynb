```python
import pygmt
import pandas as pd
import glob
from pathlib import Path
```

```python
files = glob.glob("etc/LINE-SAIL-OCEANX/*.txt")
files
```

```python
fig = pygmt.Figure()
pygmt.config(FONT_ANNOT_PRIMARY=6)
region = [90,120,-12,8]
frame = ["WSNE", "a4f"]
fig.basemap(region = region,
          frame = frame, 
          projection = "M6c")
fig.grdimage("etc/indo.nc",shading="+d",cmap="etopo1")
fig.coast(region = region, 
          projection = "M6c",
          shorelines = "0.5")

text = []
text_position_lat = []
text_position_lon = []
# read data and plot lines
for file in files:
    text.append(Path(file).stem)
    df = pd.read_csv(file, delim_whitespace=True, names=["lon","lat"])
    text_position_lat.append(df.lat[0])
    text_position_lon.append(df.lon[0])
    fig.plot(data=df, pen="0.3p,white")

#plot text
fig.text(x=text_position_lon[0]+1,y=text_position_lat[0]+0.4,text=text[0],font="3p,Helvetica",fill="white",clearance="+tO")
fig.text(x=text_position_lon[1]+1,y=text_position_lat[1]-2.7,text=text[1],font="3p,Helvetica",fill="white",clearance="+tO")
fig.text(x=text_position_lon[2]+1.5,y=text_position_lat[2]-2.5,text=text[2],font="3p,Helvetica",fill="white",clearance="+tO")
fig.text(x=text_position_lon[3]+1.5,y=text_position_lat[3]+0.5,text=text[3],font="3p,Helvetica",fill="white",clearance="+tO")
fig.text(x=text_position_lon[4]-2.7,y=text_position_lat[4]+0.5,text=text[4],font="3p,Helvetica",fill="white",clearance="+tO")
fig.text(x=text_position_lon[5]-3.5,y=text_position_lat[5],text=text[5],font="3p,Helvetica",fill="white",clearance="+tO")
fig.text(x=text_position_lon[6]-2.3,y=text_position_lat[6]-2,text=text[6],font="3p,Helvetica",fill="white",clearance="+tO")
fig.text(x=text_position_lon[7]-3,y=text_position_lat[7]-2,text=text[7],font="3p,Helvetica",fill="white",clearance="+tO")
fig.text(x=text_position_lon[8]-2,y=text_position_lat[8]-2.4,text=text[8],font="3p,Helvetica",fill="white",clearance="+tO")
fig.text(x=text_position_lon[9]-2.5,y=text_position_lat[9]-1.7,text=text[9],font="3p,Helvetica",fill="white",clearance="+tO")
fig.text(x=text_position_lon[10]+2,y=text_position_lat[10]+1,text=text[10],font="3p,Helvetica",fill="white",clearance="+tO")
fig.text(x=text_position_lon[11]+2,y=text_position_lat[11]-2.5,text=text[11],font="3p,Helvetica",fill="white",clearance="+tO")
fig.text(x=text_position_lon[12]+1.3,y=text_position_lat[12]+0.8,text=text[12],font="3p,Helvetica",fill="white",clearance="+tO")
fig.text(x=text_position_lon[13]+0.7,y=text_position_lat[13]-2.4,text=text[13],font="3p,Helvetica",fill="white",clearance="+tO")
fig.savefig("line_sail_oceanx.png", dpi=1200)
```
